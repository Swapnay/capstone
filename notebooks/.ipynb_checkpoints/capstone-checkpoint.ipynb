```python
import os
os.environ["PYSPARK_SUBMIT_ARGS"] = "--packages mysql:mysql-connector-java:8.0.22 pyspark-shell"

from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext


spark = SparkSession.builder.appName('Extract')\
                            .config("spark.ui.port", "4041")
                            .getOrCreate()

def load_from_db(spark, table_name):
    host ='spark_mysql'
    database = 'covid_economy_impact'
    port =3306
    url='jdbc:mysql://{}:{}/{}'.format(host,port, database)
    park_df = spark.read.format('jdbc').options(
            url=url,
            dbtable=table_name,
            driver= 'com.mysql.jdbc.Driver',
            user='root',
            password='Mi4man11').load()
    return park_df
```

```python

from pyspark.sql.functions import udf
from pyspark.sql.types import DateType
from datetime import datetime

table_name = 'stocks_daily_monthly_return_rate'
table_name = '(SELECT symbol,category_name,month,monthly_return_rate, year from stocks_daily_monthly_return_rate where year=2020 group by symbol,category_name, year,month,monthly_return_rate order by symbol,year,month ) as c'
table_name = '(SELECT industry_type,unemployment_rate,submission_date FROM unemployment_by_industry_fact WHERE year(submission_date) in (2020,2021) ORDER BY submission_date) as un'
park_df = load_from_db(spark,table_name)
covid_query ='(SELECT * FROM  covid_monthly_rate WHERE  country=\'USA\' ORDER BY year, month) as covid'
covid_monthly = load_from_db(spark,covid_query)

prof_bui_df = park_df.filter(park_df.industry_type=='professional business')

prof_bui_df_pandas = prof_bui_df.select('unemployment_rate').toPandas()['unemployment_rate'].tolist()
date_sub = prof_bui_df.select('submission_date').toPandas()['submission_date'].tolist()

edu_health_df = park_df.filter(park_df.industry_type=='education health care')\
                        .select('unemployment_rate')\
                        .toPandas()['unemployment_rate']\
                        .tolist()

manufacture_df = park_df.filter(park_df.industry_type=='manufacturing')\
                        .select('unemployment_rate')\
                        .toPandas()['unemployment_rate']\
                        .tolist()

agriculture_df = park_df.filter(park_df.industry_type=='agriculture related')\
                        .select('unemployment_rate')\
                        .toPandas()['unemployment_rate']\
                        .tolist()

construction = park_df.filter(park_df.industry_type=='construction')\
                      .select('unemployment_rate')\
                      .toPandas()['unemployment_rate']\
                      .tolist()

transportation = park_df.filter(park_df.industry_type=='transportation and utilities')\
                        .select('unemployment_rate')\
                        .toPandas()['unemployment_rate']\
                        .tolist()

leisure = park_df.filter(park_df.industry_type=='leisure hospitality')\
                .select('unemployment_rate')\
                .toPandas()['unemployment_rate']\
                .tolist()
def build_date(m,y):
    return datetime.now().replace(day=1, month=int(m),year=int(y))
date_udf =udf(lambda m, y:build_date(m,y) , DateType())        
covid_new = covid_monthly.withColumn('submission_date',date_udf(covid_monthly.month,covid_monthly.year))\
.select('submission_date','monthly_new_cases').toPandas()






import matplotlib.pyplot as plt
import numpy as np


labels = date_sub
#men_means = [20, 34, 30, 35, 27]
# women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
plt.rcParams['figure.figsize'] = [35, 15]
plt.rcParams.update({'font.size': 24})
fig ,ax = plt.subplots()

rects1 = ax.bar(x -0.05, prof_bui_df_pandas, 0.10, label='professional business')
rects2 = ax.bar(x + 0.05, edu_health_df, .10, label='education health care')
rects3 = ax.bar(x - 0.15, manufacture_df, .10, label='manufacturing')
rects4 = ax.bar(x + 0.15, agriculture_df, .10, label='agriculture related')
rects5 = ax.bar(x - 0.25, construction, .10, label='construction')
rects6 = ax.bar(x + 0.25, transportation, .10, label='transportation and utilities')
rects7 = ax.bar(x + 0.35, leisure, 0.10, label='leisure hospitality')

# # Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Unemployment rate')
ax.set_title('unemployment rate by sector')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)


ax2 = ax.twinx()
ax2.set_ylabel('Covid cases - Monthly - in Millions')

ax2.plot( covid_new["monthly_new_cases"],linewidth=6)

fig.tight_layout()

plt.show()
```

```python
import seaborn as sns
import pandas as pd
covid_query = '(select month, year,symbol,max(monthly_return_rate) as monthly_return_rate,category_name from stocks_daily_monthly_return_rate '\
'where category_name=\'Information Technology\' and year in (2020) '\
'group by symbol,year,month,category_name order by symbol,year,month) as sec'
symbol_query = '(select symbol from stocks_dim where category_name = \'Information Technology\') as st'
covid_monthly = load_from_db(spark,covid_query)
covid_pandas = covid_monthly.select('symbol','monthly_return_rate').toPandas()
covid_pandas['monthly_return_rate'] =covid_pandas['monthly_return_rate']+1
print(covid_pandas)
covid_pandas=covid_pandas.groupby('symbol', as_index=True).prod()
print(covid_pandas)
covid_pandas['monthly_return_rate']=covid_pandas['monthly_return_rate']-1
stock_symbols =load_from_db(spark,symbol_query)

stocks_list=stock_symbols.toPandas()['symbol'].tolist()
sns.scatterplot(data = covid_pandas, x = covid_pandas.index, y = "monthly_return_rate", hue = "symbol", size = covid_pandas.index)
plt.show()
```
