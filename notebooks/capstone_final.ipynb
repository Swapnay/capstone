```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [35, 15]
plt.rcParams.update({'font.size': 24})
fig ,ax = plt.subplots()

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        
def assign_ax(ax, x, labels, ylabel, title):
    # # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel(ylabel)
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

labels = date_sub

x = np.arange(len(labels))  # the label locations
width = 0.20  # the width of the bars

rects1 = ax.bar(x -0.05, af_df, 0.10, label='african american')
rects2 = ax.bar(x + 0.05, white_df, .10, label='white')
rects3 = ax.bar(x - 0.15, hisp_df, .10, label='hispanic')
rects4 = ax.bar(x + 0.15, asian, .10, label='asian')

assign_ax(ax, x, labels, 'Unemployment rate', 'Unemployment rate by Race')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

ax2 = ax.twinx()
ax2.set_ylabel('Covid deaths - Monthly ')

ax2.plot(covid_deaths,linewidth=6)

fig.tight_layout()

plt.show()
```

```python
print("HOUSING")

from pyspark.sql.types import StringType
import pyspark.sql.functions as F

r = requests.get('http://capn1e-api.westus.azurecontainer.io:8080//api/covid/usa/states?states=CA,TX,FL,NY' , auth=('admin', 'admin'))
response = r.json()
jsonrdd = spark.sparkContext.parallelize(response)
park_df = spark.read.json(jsonrdd)
park_df.show()
world_df = park_df.filter(park_df.year==2020).sort("year","month")#.select('totalCases').toPandas()['totalCases'].tolist()
ca_df=world_df.filter(world_df.state=='CA').select('monthlyNewCases').toPandas()['monthlyNewCases'].tolist()[:10]
tx_df=world_df.filter(world_df.state=='TX').select('monthlyNewCases').toPandas()['monthlyNewCases'].tolist()[:10]
fl_df=world_df.filter(world_df.state=='FL').select('monthlyNewCases').toPandas()['monthlyNewCases'].tolist()[:10]
il_df=world_df.filter(world_df.state=='NY').select('monthlyNewCases').toPandas()['monthlyNewCases'].tolist()[:10]


r = requests.get('http://capstone-api.westus.azurecontainer.io:8080/api/housing/inventory/states?states=CA,TX,FL,NY' , auth=('admin', 'admin'))
response = r.json()
jsonrdd = spark.sparkContext.parallelize(response)
park_df = spark.read.json(jsonrdd)
date_sub=['1-2019/2020','2-2019/2020','3-2019/2020','4-2019/2020','5-2019/2020','6-2019/2020','7-2019/2020',\
          '8-2019/2020','9-2019/2020','10-2019/2020']
park_2019 =park_df.filter((park_df.year==2019) & (park_df.inventoryType=='days_to_pending')).sort("year","month")#.select("variableName","year","month",'unemployedRate')
park_2020 =park_df.filter((park_df.year==2020) & (park_df.inventoryType=='days_to_pending')).sort("year","month")
#world_df = park_df.filter(park_df.year>2019).sort("year","month")
ca_days_df_2019=park_2019.filter(park_2019.state=='CA')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
ca_days_df_2020=park_2020.filter(park_2020.state=='CA')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
tx_df_2019=park_2019.filter(park_2019.state=='TX')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
tx_df_2020=park_2020.filter(park_2020.state=='TX')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
fl_df_2019=park_2019.filter(park_2019.state=='FL')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
fl_df_2020=park_2020.filter(park_2020.state=='FL')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
il_df_2019=park_2019.filter(park_2019.state=='NY')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]
il_df_2020=park_2020.filter(park_2020.state=='NY')\
                    .select('days')\
                    .toPandas()['days']\
                    .tolist()[:10]

```

```python
print("HOUSING DAYS TO PENDING IN STATES WITH HIGHEST MONTHLY COVID CASES ")
labels = date_sub
plt.rcParams['figure.figsize'] = [45, 20]
plt.rcParams.update({'font.size': 24})

#  def paint(labels,plt,*args,**kwargs):
# #     #for i, k in kwargs.items():
fig ,ax = plt.subplots()
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
#fig ,ax = plt.subplots()
print(il_df_2019)
print(il_df_2020)

rects1 = ax.bar(x -0.05, ca_days_df_2019, .10, label='CA',color='fuchsia')
rects2 = ax.bar(x + 0.05, ca_days_df_2020, .10, label='CA',color='fuchsia')
rects3 = ax.bar(x - 0.15, tx_df_2019, .10, label='TX',color='green')
rects4 = ax.bar(x + 0.15, tx_df_2020, .10, label='TX',color='green')
rects5 = ax.bar(x - 0.25, fl_df_2019, .10, label='FL',color='orange')
rects6 = ax.bar(x + 0.25, fl_df_2020, .10, label='FL',color='orange')
rects7 = ax.bar(x - 0.35, il_df_2019, .10, label='NY',color='turquoise')
rects8 = ax.bar(x + 0.35, il_df_2020, .10, label='NY',color='turquoise')
# rects7 = ax.bar(x - 0.35, MI_df_2019, .10, label='MICHIGAN',color='orange')
# rects8 = ax.bar(x + 0.35, MI_df_2020, .10, label='MICHIGAN',color='orange')
# rects9 = ax.bar(x - 0.45, ny_df_2019, .10, label='New York',color='turquoise')
# rects10 = ax.bar(x + 0.45, ny_df_2020, .10, label='New York',color='turquoise')

# rects7 = ax.bar(x + 0.35, leisure, 0.10, label='leisure hospitality')

# # Add some text for labels, title and custom x-axis tick labels, etc.
assign_ax(ax, x, labels, 'Number Of Days To Sle Pending', 'Housing Sale Pending Days - COVID Cases')

autolabel(rects1) 
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)
autolabel(rects8)
# autolabel(rects9)
# autolabel(rects10)


ax2 = ax.twinx()
ax2.set_ylabel('COVID Cases in 100000s')
# Plot the relative temperature in red
ax2.plot(ca_df,label="CA",linewidth=10,color='fuchsia')
ax2.plot(tx_df,label="TX",linewidth=10,color='green')
ax2.plot(fl_df,label="FL",linewidth=10,color='orange')
ax2.plot(il_df,label="NY",linewidth=10,color='turquoise')


fig.tight_layout()
#plt.legend()


plt.show()
```
